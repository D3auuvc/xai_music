## Preprocessing data

- one hot encoding


```python
# one hot encoding

ohe_df = Albums_Cafe_Paradiso_06.drop('tags', 1).join(
    Albums_Cafe_Paradiso_06.tags.str.join('|').str.get_dummies())


ohe_df.drop(ohe_df.columns[[1]], axis=1, inplace=True)
ohe_df.groupby(['key', 'tempo']).sum()
```

```python
group_ohe_df = ohe_df.groupby(['key']).sum()
group_ohe_df
```

```python
from matplotlib import ticker
import matplotlib.pyplot as plt


def plot_result(group_ohe_df):
    fig, ax = plt.subplots()
    fig.set_figwidth(20)

    arr = group_ohe_df.to_numpy()
    key, tag = arr.shape

    x_lables_lits = [col for col in group_ohe_df.columns]
    y_lables_lits = group_ohe_df.index.tolist()

    ax.matshow(arr, cmap=plt.cm.Blues)
    ax.set_xlabel('tag')
    ax.set_ylabel(group_ohe_df.index.name)
    ax.xaxis.set_ticklabels([''] + x_lables_lits, rotation=90)
    ax.yaxis.set_ticklabels([''] + y_lables_lits)
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax.yaxis.set_major_locator(ticker.MultipleLocator(1))

    for i in range(tag):
        for j in range(key):
            c = arr[j, i]
            ax.text(i, j, str(c), va='center', ha='center')

    plt.show()


plot_result(group_ohe_df)
```

```python
def plot_multi_result(groupby_dict, size):
    fig, ax = plt.subplots(size, 1, figsize=(100, 100), gridspec_kw={
        'wspace': 0.4,
        'hspace': 0.8})

    idx = 0
    for instrumrnt, group_ohe_df in groupby_dict.items():
        arr = group_ohe_df.to_numpy()
        key, tag = arr.shape
        groupby = group_ohe_df.index.name
        fig.suptitle(f'Group by {groupby}')

        x_lables_lits = [col for col in group_ohe_df.columns]
        y_lables_lits = group_ohe_df.index.tolist()

        ax[idx].matshow(arr, cmap=plt.cm.Blues)

        ax[idx].set_xlabel('tag')
        ax[idx].set_title(instrumrnt)
        ax[idx].set_ylabel(groupby)

        ax[idx].xaxis.set_ticklabels([''] + x_lables_lits, rotation=90)
        ax[idx].yaxis.set_ticklabels([''] + y_lables_lits)
        ax[idx].xaxis.set_major_locator(ticker.MultipleLocator(1))
        ax[idx].yaxis.set_major_locator(ticker.MultipleLocator(1))

        for i in range(tag):
            for j in range(key):
                c = arr[j, i]
                ax[idx].text(i, j, str(c), va='center', ha='center')
        idx += 1

    plt.show()
```

## Categorize data by music
We categorize tag result by `music`, then analyzing:
1. The influence of each `key` in different `tempo`.
2. The influence of each `tempo` in different `key`.

### 1. The influence of each `key` in different `tempo`

Although we pick original audio file randomly, the pattern of result is similar in each round of pick original audio file.

- Cha Cha
For example, in Cha Cha, when key transfer to lower semitone(-1, -2), the model detect vocal from female to male. When key transfer to higher semitone(1, 2), model satrted to detect indian genre. In Jive, When key transfer to lower semitone(-1, -2), the model loss pop genre. When key transfer to higher semitone(1, 2), model satrted to get pop genre more. Viennese Waltz has stable result. Some cases(Tango, or Waltz) start to detect flute tag from female vocal when increase key.

```python
# By key
groupby_key_dict = dict()

for key, value in music_dict.items():
    ohe_df = value.drop('tags', 1).join(
        value.tags.str.join('|').str.get_dummies())
    ohe_df.drop(ohe_df.columns[[1]], axis=1, inplace=True)
    groupby_key_dict[key] = ohe_df.groupby(['key']).sum()


plot_multi_result(groupby_key_dict, len(groupby_key_dict))
```

### 2. The influence of each `tempo` in different `key`

```python
# By tempo
groupby_tempo_dict = dict()

for key, value in music_dict.items():
    ohe_df = value.drop('tags', 1).join(
        value.tags.str.join('|').str.get_dummies())
    ohe_df.drop(ohe_df.columns[[0]], axis=1, inplace=True)
    groupby_tempo_dict[key] = ohe_df.groupby(['tempo']).sum()


plot_multi_result(groupby_tempo_dict, len(groupby_tempo_dict))
```

## Categorize data by key
We categorize tag result by `key`, then analyzing:
1. The influence of each `music` in different `tempo`.
2. The influence of each `tempo` in different `music`.


The result showed that similar distribution. Hence, the `music` and `tempo` do not influence the tag result in every `key`.

```python
# Categorize function
# features: music, tempo

def categorize(data, feature):
    categorized_df = data[feature].unique().tolist()
    categorized_dict = dict()
    for i in categorized_df:
        categorized_dict[i] = data.loc[data[feature] == i]

    return categorized_dict


categorized_dict = categorize(data, 'key')
categorized_dict['2']
```

```python
# By music
groupby_music_dict = dict()

for key, value in categorized_dict.items():
    ohe_df = value.drop('tags', 1).join(
        value.tags.str.join('|').str.get_dummies())
    # ohe_df.drop(ohe_df.columns[[0]], axis=1, inplace=True)
    groupby_music_dict[key] = ohe_df.groupby(['music']).sum()

# groupby_music_dict['original']
plot_multi_result(groupby_music_dict, len(groupby_music_dict))
```

```python
# By tempo
groupby_tempo_dict = dict()

for key, value in categorized_dict.items():
    ohe_df = value.drop('tags', 1).join(
        value.tags.str.join('|').str.get_dummies())
    # ohe_df.drop(ohe_df.columns[[0]], axis=1, inplace=True)
    groupby_tempo_dict[key] = ohe_df.groupby(['tempo']).sum()

# groupby_tempo_dict['original']
plot_multi_result(groupby_tempo_dict, len(groupby_tempo_dict))
```
